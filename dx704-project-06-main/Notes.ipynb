To calculate the expected discomfort of a policy that's always "do nothing",  use *value iteration*.  `expected discomfort =  -1 * expected reward`.

2.  **Define a fixed policy**: The policy is to always take the "do-nothing" action, so  use the `value_iteration_fixed` function to evaluate it.

3.  **Run value iteration**: Use the helper function `value_iteration_fixed` with the "do-nothing" R and P.

4.  **Calculate discomfort**: The returned value function will be the expected cumulative *reward* for each state. Convert these values to discomfort by multiplying them by -1, since discomfort is the negative of the reward.

The `value_iteration_fixed` function is designed for a single, fixed policy. It takes a reward vector and a transition matrix.
The reward vector `R` is the immediate reward for being in each state, and the transition matrix `P` contains the probabilities of moving from one state to another under the "do-nothing" action.

Run `value_iteration_fixed(R, P, gamma = 0.99)`; the output is the value function, which represents the *expected cumulative rewards*. To get the expected discomfort,  multiply these values by **-1**, b/c discomfort is the negative of the reward.

For Part 2, the goal is to compute the optimal treatment plan that minimizes patient discomfort, which is equivalent to maximizing the expected cumulative rewards. The provided notebook includes the `value_iteration_multiple` and `policy_iteration_baseline` functions, which are designed for this purpose.

2.  **Run the algorithm**: Use the `value_iteration_multiple` function to find the optimal value function. Value iteration works by iteratively applying the Bellman optimality update, which finds the value function that maximizes expected cumulative rewards across all available actions. 3.  **Extract the optimal policy**: Once the optimal value function has been computed, you can find the optimal policy. Then, take the `argmax` over the actions for each state to find the optimal action.
4.  **Save the results**: Save the optimal actions for each state to a file named `minimum-discomfort-actions.tsv` with the columns `state` and `action`.
